```python
import random
from smolagents import tool, CodeAgent, OpenAIModel
```

```python
from pathlib import Path
from dotenv import load_dotenv
```

```python
# Carregando as variáveis de ambiente
ENV_PATH = Path('../.env')
load_dotenv(dotenv_path=ENV_PATH)
```

```python
@tool
def throw_dice_tool() -> int:
    'You roll a die and get a persons name.'
    DICE = {
        'ONE': 'Luiz',
        'TWO': 'Two',
        'THREE': 'Nuna',
        'FOUR': 'Henrique',
        'FIVE': 'Egito',
        'SIX': 'Luiz Henrique'
    }
    number = random.choice(list(DICE.keys()))
    return f'O numero do dado foi: {number}, e o nome escolhido foi: {DICE[number]}'
```

```python
model = OpenAIModel(
    model_id="gpt-4o",
)

agent = CodeAgent(tools=[throw_dice_tool], model=model, verbosity_level=0)

result = agent.run("Jogue o dado e me diga qual nome da pessoa e qual numero ele caiu.")
print(result)
```
