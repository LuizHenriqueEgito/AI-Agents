```python
from smolagents import CodeAgent, OpenAIModel
```

```python
model = OpenAIModel(
    model_id="gpt-4o",     # ou outro modelo da OpenAI
)
```

```python
agent = CodeAgent(
    tools=[],            # sem ferramentas extras para este exemplo
    model=model,
    verbosity_level=0
)
```

```python
agent.run("Olá como vai você?")
```
