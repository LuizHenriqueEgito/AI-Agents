## Buscas Web

```python
from smolagents import CodeAgent, OpenAIModel
from smolagents import ToolCallingAgent, WebSearchTool
```

```python
model = OpenAIModel(
    model_id="gpt-4o",     # ou outro modelo da OpenAI
)
```

```python
agent = ToolCallingAgent(
    tools=[WebSearchTool()],
    model=model,
    verbosity_level=0
)
```

```python
prompt = '''
Quais times estão na final da Libertadores de 2025 e quando será o jogo?
'''
response = agent.run(prompt)
```
