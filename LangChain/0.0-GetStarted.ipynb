## Busca Web

```python
from langchain_community.tools import DuckDuckGoSearchRun
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

search = DuckDuckGoSearchRun(region="br-pt", max_results=5)
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

prompt = PromptTemplate.from_template("""
Resultados da pesquisa web:
{results}

Pergunta relacionada:
{query}

Responda de forma clara e objetiva, usando os dados da pesquisa.
""")

chain = (
    {
        "query": lambda x: x,
        "results": lambda x: search.run(x)
    }
    | prompt
    | llm
    | StrOutputParser()
)
prompt = 'Quais times estão na final da Libertadores de 2025 e qando será o jogo?'
print(chain.invoke(prompt))
```
