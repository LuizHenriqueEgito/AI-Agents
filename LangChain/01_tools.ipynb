```python
import random
from pathlib import Path
from dotenv import load_dotenv
```

```python
from langchain.tools import tool
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain.agents import create_agent
```

```python
# Carregando as variáveis de ambiente
ENV_PATH = Path('../.env')
load_dotenv(dotenv_path=ENV_PATH)
```

```python
@tool
def throw_dice_tool() -> int:
    'You roll a die and get a persons name.'
    DICE = {
        'ONE': 'Luiz',
        'TWO': 'Two',
        'THREE': 'Nuna',
        'FOUR': 'Henrique',
        'FIVE': 'Egito',
        'SIX': 'Luiz Henrique'
    }
    number = random.choice(list(DICE.keys()))
    return f'O numero do dado foi: {number}, e o nome escolhido foi: {DICE[number]}'

print(throw_dice_tool.name)

print(throw_dice_tool.description)
```
