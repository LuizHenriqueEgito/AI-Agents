```python
from datetime import datetime
from crewai import Agent
from crewai.tools import tool

@tool
def today_date() -> str:
    """Retorna a data atual no formato YYYY-MM-DD."""
    return datetime.now().strftime("%Y-%m-%d")
```

```python
@tool
def web_search(query: str) -> str:
    '''Retorna o resultado da pesquisa pelo tópico.'''
    search = DuckDuckGoSearchRun(
        region="br-pt",
        safesearch="moderate"
    )
    return search.run(query)

agent = Agent(
    role="Pesquisador web",
    goal="Responder perguntas sobre a pesquisa do usuario usando a pesquisa web",
    backstory="Você sempre sabe informações atuais pois tem acesso a web",
    tools=[web_search],     # <-- adicionando a função como tool
    model="gpt-4o",
    respects_tools=True
)

task = Task(
    description="Use a pesquisa web para responder a pergunta: {topic}",
    expected_output='Com base na sua pesquisa responda ao usuario',
    agent=agent,
)

crew = Crew(
    agents=[agent],
    tasks=[task],
)

topic = 'Quais times estão na final da Libertadores de 2025 e quando será o jogo?'
result = crew.kickoff(inputs={'topic': topic})
```
