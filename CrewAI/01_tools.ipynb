```python
import random
from pathlib import Path
from dotenv import load_dotenv
from datetime import datetime

from crewai import Agent, Task, Crew
from crewai.tools import tool
```

```python
# Carregando as variáveis de ambiente
ENV_PATH = Path('../.env')
load_dotenv(dotenv_path=ENV_PATH)
```

```python
@tool
def throw_dice_tool() -> int:
    'You roll a die and get a persons name.'
    DICE = {
        'ONE': 'Luiz',
        'TWO': 'Two',
        'THREE': 'Nuna',
        'FOUR': 'Henrique',
        'FIVE': 'Egito',
        'SIXTH': 'Luiz Henrique'
    }
    number = random.choice(list(DICE.keys()))
    return f'O numero do dado foi: {number}, e o nome escolhido foi: {DICE[number]}'
```

```python
agent = Agent(
    role="Pesquisador web",
    goal="Responder perguntas sobre a pesquisa do usuario usando a pesquisa web",
    backstory="Você sempre sabe informações atuais pois tem acesso a web",
    tools=[throw_dice_tool],
    model="gpt-4o",
    respects_tools=True
)

task = Task(
    description="""
    Use obrigatoriamente a tool throw_dice_tool para lançar o dado.
    Não invente a resposta.
    A resposta deve vir diretamente do resultado da tool.
    """,
    expected_output="O resultado do lançamento do dado retornado pela tool",
    agent=agent,
)

crew = Crew(
    agents=[agent],
    tasks=[task],
)
```

```python
result = crew.kickoff()
```
