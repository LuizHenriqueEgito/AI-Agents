```python
import agno
from pathlib import Path
from dotenv import load_dotenv

from agno.agent import Agent, RunOutput
from agno.models.openai import OpenAIChat
```

```python
# Carregando as variáveis de ambiente
ENV_PATH = Path('../.env')
load_dotenv(dotenv_path=ENV_PATH)
```

```python
agent = Agent(model=OpenAIChat(id="gpt-5-mini"), markdown=True)
run_response = agent.run("Olá como vai você?")
```

```python
print(run_response.content)
```

## Buscas Web

```python
from agno.tools.duckduckgo import DuckDuckGoTools
```

```python
agent = Agent(model=OpenAIChat(id="gpt-5-mini"), tools=[DuckDuckGoTools()])
prompt = '''
Quais times estão na final da Libertadores de 2025 e quando será o jogo?
'''
run_response = agent.run(prompt)
```

```python
print(run_response.content)
```
