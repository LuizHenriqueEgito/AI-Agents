```python
import random
import agno
from pathlib import Path
from dotenv import load_dotenv

from agno.agent import Agent, RunOutput
from agno.models.openai import OpenAIResponses
from agno.tools import tool
```

```python
# Carregando as variáveis de ambiente
ENV_PATH = Path('../.env')
load_dotenv(dotenv_path=ENV_PATH)
```

# Fazendo uma tool dummy

```python
@tool
def throw_dice_tool() -> int:
    'You roll a die and get a persons name.'
    DICE = {
        'ONE': 'Luiz',
        'TWO': 'Two',
        'THREE': 'Nuna',
        'FOUR': 'Henrique',
        'FIVE': 'Egito',
        'SIXTH': 'Luiz Henrique'
    }
    number = random.choice(list(DICE.keys()))
    return f'O numero do dado foi: {number}, e o nome escolhido foi: {DICE[number]}'
```

```python
agent = Agent(
    model=OpenAIResponses(id="gpt-4.1-mini"),
    tools=[throw_dice_tool],
)
prompt = 'Lance um dado e me o seu número e o nome que o dado escolheu por favor.'

response = agent.run(prompt)
print(response.content)
```
